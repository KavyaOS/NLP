```python
import pandas as pd
```

```python
# Y3test.csv includes all csv files
df = pd.read_csv('Y3test.csv', error_bad_lines=False)
df.columns = ['Text']

df
```

```python
df[['ID', 'Word']] = df['Text'].str.split('\t', 1, expand=True)
df = df.drop('Text', 1)

df
```

```python
values = ['70', '69', '43', '42', '40', '78', '52', '39', '79', '67', '51', '48', '66', '60', '61', '129', '72']

df = df.loc[df['ID'].isin(values)]

df
```

```python
merge = (df['ID'] != df['ID'].shift()).cumsum().rename('grp')
x = df.groupby(['ID', merge], sort=False)['Word']\
      .agg(' '.join).reset_index().drop('grp', axis=1)

x['ID'] = '<' + x['ID'].astype(str) + '>'

x
```

```python
x["Text"] = x["ID"] + ' ' + x["Word"]

x = x.drop('ID', 1)
x = x.drop('Word', 1)

x
```

```python
x.to_csv("cleaned_data_final.txt", header=None, index=None, sep='\n', mode='w')
```
